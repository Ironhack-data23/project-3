```python
def companies_ubi(name, city, latitude, longitude):
    filtering= {"number_of_employees":{"$gte":90,"$lte":140},"category_code":"games_video"}
    projection= {"_id":0,"name":1,"number_of_employees":1,"offices.city":1,"category_code":1,"offices.latitude":1,"offices.longitude":1,"outcode":1}
    usa_companies = list(c.find(filtering, projection).sort("number_of_employees", 1))
    name = []
    latitude = []
    longitude = []
    city = []
    country = []
    for i in usa_companies:
        offices = i["offices"]

        for j in offices:
            # Check if the latitude and longitude for the current office are not None
            if j["city"] is not None and j["latitude"] is not None and j["longitude"] is not None:
                name.append(i["name"]) 
                city.append(j["city"])
                country.append(j["country_code"])
                latitude.append(j["latitude"])
                longitude.append(j["longitude"])

    return name,city,latitude,longitude
```

```python
def export_to_json(usa_companies, output_directory, output_file):
    try:
        full_file_path = os.path.join(output_directory, output_file)
        os.makedirs(output_directory, exist_ok=True)
        with open(full_file_path, 'w') as json_file:
            json.dump(usa_companies, json_file, indent=4)
        print(f"Data exported to '{full_file_path}'")
        return True
    except Exception as e:
        print(f"Error: {e}")
        return False

output_directory = 'data/'
output_file_name = 'usa_companiess.json'

result = export_to_json(usa_companies, output_directory, output_file_name)
```
